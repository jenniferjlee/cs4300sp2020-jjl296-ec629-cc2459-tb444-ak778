```python
from bs4 import BeautifulSoup
from requests import get
import time
import pandas as pd
from random import randint
```

```python
df = pd.DataFrame(columns = ['source', 'title', 'subtitle', 'date', 'link', 'transcript'])
```

```python
df
```

```python
for page_num in range(9, 11):
    page_url = 'https://www.newser.com/tag/68850/' + str(page_num) + '/uplifting-news.html'
    page_soup = BeautifulSoup(get(page_url).text, 'html.parser')
    
    news_headlines = [h.get_text() for h in page_soup.findAll("h2")]
    news_sub_headlines = [h.get_text() for h in page_soup.findAll("h3")]
    links = [('https://www.newser.com' + l.find("a")['href']) for l in page_soup.findAll("div", attrs={'class': 'storyListContainer'})]
    
    for ind in range(len(links)):
        print(ind)
        time.sleep(randint(1, 5))
        article_url = links[ind]
        sub_page_soup = BeautifulSoup(get(article_url).text, 'html.parser')
        
        sub_page_content = ''
        for p in sub_page_soup.findAll("p", attrs = {'class': 'storyParagraph'}):
            sub_page_content += p.get_text()
        sub_page_content = sub_page_content.replace('\n', '')
        sub_page_content = sub_page_content.replace('\r', '')
        sub_page_content = sub_page_content.replace("\\", '')
        sub_page_content = sub_page_content.replace('(Newser)', '')
        sub_page_content = sub_page_content.replace('(Read more uplifting news stories.)', '')
        
        sub_page_date = sub_page_soup.find('span', attrs = {'id': 'spanPublishedDate'}).get_text().replace('Posted ', '')
        
        df = df.append({
            'source': 'Newser',
            'title': news_headlines[ind],
            'subtitle': news_sub_headlines[ind],
            'date': sub_page_date,
            'link': article_url,
            'transcript': sub_page_content
        }, ignore_index=True)
```

```python
df
```

```python
df = df.drop_duplicates().reset_index()
```

```python
df = df.drop(['index', 'level_0'], axis = 1)
```

```python
df.to_json('newser_draft1.json', orient='records')
```
